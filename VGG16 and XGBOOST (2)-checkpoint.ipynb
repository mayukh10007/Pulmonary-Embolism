```python
import numpy as np 
import matplotlib.pyplot as plt
import glob
import cv2

from keras.models import Model, Sequential
from keras.layers import Dense, Flatten, Conv2D, MaxPooling2D
from tensorflow.keras.layers import BatchNormalization
import os
import seaborn as sns
from keras.applications.vgg16 import VGG16
import numpy as np
```

```python
# Read input images and assign labels based on folder names
print(os.listdir(r"C:\Users\ACER\Desktop\Medical-Image-Classification-main\Dataset\PE_Split"))
```

```python
SIZE = 256  # Resize images
```

```python
# Capture training data and labels into respective lists
train_images = []
train_labels = []

train_dir = r"C:\Users\ACER\Desktop\Medical-Image-Classification-main\Dataset\PE_Split\train"

for label in os.listdir(train_dir):
    label_path = os.path.join(train_dir, label)
    print(label)
    for img_path in os.listdir(label_path):
        print(img_path)
        img = cv2.imread(os.path.join(label_path, img_path), cv2.IMREAD_COLOR)
        img = cv2.resize(img, (SIZE, SIZE))
        img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        img = np.stack((img,) * 3, axis=-1)
        train_images.append(img)
        train_labels.append(label)
```

```python
len(train_images), len(train_labels)
```

```python
# Convert lists to arrays
train_images = np.stack(train_images, axis=0)
train_labels = np.array(train_labels)
```

```python
train_images.shape
```

```python
test_images = []
test_labels = []

test_dir = r"C:\Users\ACER\Desktop\Medical-Image-Classification-main\Dataset\PE_Split\validation"

for val_label in os.listdir(test_dir):
    val_label_path = os.path.join(test_dir, val_label)
    for img_path in os.listdir(val_label_path):
        img = cv2.imread(os.path.join(val_label_path, img_path), cv2.IMREAD_COLOR)
        img = cv2.resize(img, (SIZE, SIZE))
        img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        img = np.stack((img,) * 3, axis=-1)
        test_images.append(img)
        test_labels.append(val_label)
```

```python
len(test_images), len(test_labels)
```

```python
# Convert lists to arrays                
test_images = np.stack(test_images, axis=0)
test_labels = np.array(test_labels)
```

```python
test_images.shape
```

```python
plt.figure()
plt.imshow(test_images[2])
plt.axis('off')
plt.show()
```

```python
# Encode labels from text to integers
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

test_labels_encoded = le.fit_transform(test_labels)
train_labels_encoded = le.fit_transform(train_labels)
```

```python
# Split data into test and train datasets (already split but assigning to meaningful convention)
x_train, y_train, x_test, y_test = train_images, train_labels_encoded, test_images, test_labels_encoded
```

```python
# Normalize pixel values to between 0 and 1
x_train = x_train.astype('float32') / 255.0
x_test = x_test.astype('float32') / 255.0

# One hot encode y values for neural network. 
# from keras.utils import to_categorical
# y_train_one_hot = to_categorical(y_train)
# y_test_one_hot = to_categorical(y_test)
```

```python
# Load model without classifier/fully connected layers
from tensorflow.keras.applications import VGG16

VGG_model = VGG16(weights='imagenet', include_top=False, input_shape=(SIZE, SIZE, 3))
```

```python
# Make loaded layers as non-trainable. This is important as we want to work with pre-trained weights
for layer in VGG_model.layers:
    layer.trainable = False
```

```python
VGG_model.summary()  #Trainable parameters will be 0
```

```python
# Now, let us use features from convolutional network for RF
feature_extractor=VGG_model.predict(x_train)
```

```python
features = feature_extractor.flatten().reshape(feature_extractor.shape[0], -1)
```

```python
X_for_training = features #This is our X input to RF
```

```python
# !pip install xgboost 
```

```python
len(X_for_training[1])
X_for_training.shape
```

```python
# XGBOOST
import xgboost as xgb
model = xgb.XGBClassifier()
model.fit(X_for_training, y_train) #For sklearn no one hot encoding
```

```python
# Send test data through same feature extractor process
X_test_feature = VGG_model.predict(x_test)
X_test_features = X_test_feature.reshape(X_test_feature.shape[0], -1)
```

```python
# Now predict using the trained RF model. 
prediction = model.predict(X_test_features)
#Inverse le transform to get original label back. 
prediction = le.inverse_transform(prediction)
```

```python
# Print overall accuracy
accuracy = (prediction == test_labels).mean()
print("Accuracy = ", accuracy)
```

```python
# Confusion Matrix - verify accuracy of each class
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(test_labels, prediction)
plt.figure(figsize=(8,6))
sns.heatmap(cm, annot=True, cmap="Blues", fmt="d", xticklabels=le.classes_, yticklabels=le.classes_)
plt.title("Confusion Matrix")
plt.xlabel("Predicted Labels")
plt.ylabel("True Labels")
plt.show()
```

```python
from sklearn.metrics import classification_report
report = classification_report(test_labels, prediction)
print(report)
```
